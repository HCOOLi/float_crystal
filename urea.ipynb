```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
plt.figure(figsize=(4, 3), dpi=460)

for i in range(0,10,1):
    Ep2=[]
    T=[]
    for j in np.arange(0.1,1.1,0.1):
        data = pd.read_csv("/home/hcooh/fastdata/ureastepheat/ureastepheat/Ep2=%gEe2e=%d/crystility"%(j,i),header=None,sep="\t",names=["T","cb","cs"])
        data = data [~ data ['cs'].isin([0])]
        Ep2.append(j)
#         plt.plot(data["T"],data["cs"])
#         plt.show()
#         os.system("pause")
        T.append(data["T"][data.diff()["cs"].idxmin()])
        print("%g"%i,"%g"%j,data["T"][data.diff()["cs"].idxmin()] )
#         print("%g"%(0.3*i+3.5*j))
    plt.plot(Ep2,T,"^-",label="Ee2e=%g"%i)
    #plt.title("stepcool: Ee2e=%g"%i)
    plt.xlabel("${Ep_2}$", fontdict={'family' : 'Times New Roman', 'size':8})
    plt.ylabel("${T_m}$", fontdict={'family' : 'Times New Roman', 'size':8})
    plt.xticks(fontproperties = 'Times New Roman', fontsize=8)
    plt.yticks(np.arange(0, 6, 1), fontproperties = 'Times New Roman', fontsize=8)
plt.title('stepheating', fontdict={'family' : 'Times New Roman', 'size':8})
plt.legend(loc='lower right', prop={'family':'Times New Roman', 'size':5})
plt.show()
#         5*Ep1()
    #     plt.plot(data["T"],data["C"])
#     plt.show()
```
